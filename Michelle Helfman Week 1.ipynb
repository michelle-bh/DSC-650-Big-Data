Employment and education numbers are shown as percentages so the states with the largest and smallest populations will have the equal representation.  The income is the average income per gender and state.<br><br>

```python
# Create data frames for all records
MVF_df = pd.read_excel('Michelle_Helfman_Male_vs_Female_EXCEL.xlsx', sheet_name='MVF')

print('1st 10 Rows of MVF Dataset')
MVF_df.head(10)
```

### Data Definitions/Column Information

 - State: state name in alphabetical order - string
 - Sex:  Gender - string
 - State_Population_PCT:  Percentage of population by state and gender - decimal
 - Labor_Participation_PCT:  Number of working age people that are employed or actively seeking employment, divided by the working-age population by state and gender (percentage) - decimal
 - Employment_PCT:  Percentage of potentially employed people who have jobs by state and gender - decimal
 - High_School_PCT:  Percentage of students who started high school in ninth grade and graduated four or five years later by state and gender - decimal
 - College_Grad_PCT:  Percentage of students who started college/university and graduated four to six years later by state and gender - decimal
 - Income:  Average annual income by state and gender - integer 

## 2.  Create a histogram or bar graph from your data.

### Employment, Income, and Education by Gender.

```python
# Employment % by state and sex

employ_hist = sns.histplot(data=MVF_df, x='Employment_PCT', 
                           bins=15, hue='Sex', multiple='dodge')
employ_hist.set_title('Employment Percentage by Gender', 
                      fontdict={'size': 15, 'weight': 'bold', 'color': 'black'})
employ_hist.set(xlabel='Employment Rate %', ylabel='# of States')

plt.show()
```

```python
# Sort the Records by Employment Rate and 
# find the highest and lowest and show State and Sex

sorted_emp = MVF_df.sort_values('Employment_PCT',ascending = False)
print('\nTop 10 Employment Rate')
print(sorted_emp[['State', 'Sex', 'Employment_PCT']].head(10))
print('\nBottom 10 Employment Rate')
print(sorted_emp[['State', 'Sex', 'Employment_PCT']].tail(10))
```

```python
# Income by state and sex

income_hist = sns.histplot(data=MVF_df, x='Income', 
                           bins=15, hue='Sex', multiple='dodge')
income_hist.set_title('Income ($s) by Gender', 
                      fontdict={'size': 15, 'weight': 'bold', 'color': 'black'})
income_hist.set(xlabel='Income $s', ylabel='# of States')

plt.show()
```

```python
# Sort the Records by Income Level and 
# find the highest and lowest and show State and Sex

sorted_income = MVF_df.sort_values('Income',ascending = False)
print('\nTop 10 Income Level')
print(sorted_income[['State', 'Sex', 'Income']].head(10))
print('\nBottom 10 Income Level')
print(sorted_income[['State', 'Sex', 'Income']].tail(10))
```

```python
# High School Graduation Rate by state and sex

hs_hist = sns.histplot(data=MVF_df, x='High_School_PCT', 
                       bins=15, hue='Sex', multiple='dodge')
hs_hist.set_title('High School Graduation by Gender', 
                      fontdict={'size': 15, 'weight': 'bold', 'color': 'black'})
hs_hist.set(xlabel='Graduation Rate %', ylabel='# of States')

plt.show()
```

```python
# Sort the Records by High School Graduation Rate and 
# find the highest and lowest and show State and Sex

sorted_hsg = MVF_df.sort_values('High_School_PCT',ascending = False)
print('\nTop 10 High School Graduation Rate')
print(sorted_hsg[['State', 'Sex', 'High_School_PCT']].head(10))
print('\nBottom 10 High School Graduation Rate')
print(sorted_hsg[['State', 'Sex', 'High_School_PCT']].tail(10))
```

## 3.  Create a boxplot from your data.

### Employment, Income, and Education by Gender.

```python
# Employment % by Gender

employ_bp = sns.boxplot(data = MVF_df, x = 'Sex', y = 'Employment_PCT', hue = 'Sex',
                        order = ['Female', 'Male'], orient = 'v', dodge=False)
employ_bp.set_title('Employment Percentage by Gender', 
                      fontdict={'size': 15, 'weight': 'bold', 'color': 'black'})
employ_bp.set(xlabel='Employment %', ylabel='Gender')

plt.show()
```

```python
# Find the mean employment % by Gender 

emp_mean = MVF_df.groupby(['Sex'])['Employment_PCT'].mean()

print('Average Employment Rate by Gender')
print(round(emp_mean,2))
```

```python
# Income by Gender

income_bp = sns.boxplot(data = MVF_df, x = 'Sex', y = 'Income', hue = 'Sex',
                        order = ['Female', 'Male'], orient = 'v', dodge=False)
income_bp.set_title('Income ($) by Gender', 
                      fontdict={'size': 15, 'weight': 'bold', 'color': 'black'})
income_bp.set(xlabel='Income ($)', ylabel='Gender')

plt.show()
```

```python
# Find the mean income % by Gender 

income_mean = MVF_df.groupby(['Sex'])['Income'].mean()

print('Average Income by Gender')
print(round(income_mean,2))
```

```python
# High School Graduation Rate by Gender

hs_grad_bp = sns.boxplot(data = MVF_df, x = 'Sex', y = 'High_School_PCT', hue = 'Sex',
                        order = ['Female', 'Male'], orient = 'v', dodge=False)
hs_grad_bp.set_title('High School Graduation by Gender', 
                      fontdict={'size': 15, 'weight': 'bold', 'color': 'black'})
hs_grad_bp.set(xlabel='High School Graduation %', ylabel='Gender')

plt.show()
```

```python
# Find the mean High School Graduation % by Gender 

hs_mean = MVF_df.groupby(['Sex'])['High_School_PCT'].mean()

print('Average High School Graduation Rate by Gender')
print(round(hs_mean,2))
```

## 4.  Create a bivariate plot from your data.

### Cross Comparing Gender Employment, Income, and Education Using Scatter Plots.  Comparing 2 Columns.

```python
# Employment and Income for both Males and Females 

empinc_scatter = sns.scatterplot(data = MVF_df, x='Employment_PCT', 
                                 y='Income', hue='Sex')
empinc_scatter.set_title(
    "Employment & Annual Income ($) by Gender")
empinc_scatter.set(ylabel="Annual Income", xlabel="Employment %")

plt.show()
```

```python
# High School Graduation and Employment for both Males and Females

emphs_scatter = sns.scatterplot(data=MVF_df, x='High_School_PCT', 
                                 y='Employment_PCT', hue='Sex')
emphs_scatter.set_title(
    "High School Graduation & Employment % by Gender")
emphs_scatter.set(ylabel="Employment %", xlabel="High School Graduation %")

plt.show()
```

```python
# High School Graduation and Income for both Males and Females

hsinc_scatter = sns.scatterplot(data=MVF_df, x='High_School_PCT', 
                                 y='Income', hue='Sex')
hsinc_scatter.set_title(
    "High School Graduation & Income ($) by Gender")
hsinc_scatter.set(ylabel="Income ($)", xlabel="High School Graduation %")

plt.show()
```

## 5.  Additional Visualizations

### Population and College Graduation

```python
# The average of all the percentages in the United States 
# separated by gender

# Create pie chart
MVF_df.groupby(['Sex']).mean().plot(kind='pie', 
            y='State_Population_PCT', autopct='%1.1f%%', colors=cust_colors)
plt.title("US Population % by Males and Females", bbox={'facecolor':'1.0', 'pad':5})

plt.show()
```

```python
# College Graduation and Income for both Males and Females

cginc_scatter = sns.scatterplot(data=MVF_df, x='College_Grad_PCT', 
                                 y='Income', hue='Sex')
cginc_scatter.set_title(
    "College Graduation & Income ($) by Gender")
cginc_scatter.set(ylabel="Income ($)", xlabel="College Graduation %")

plt.show()
```
